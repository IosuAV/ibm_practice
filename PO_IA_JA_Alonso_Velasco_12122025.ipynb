#### Ejercicio 5.2
Separa el dataset provisto en conjunto de entrenamiento y de test, dando un 40% de los datos al conjunto de test. La variable objetivo corresponde a la columna "species"

```python
# Resuelve aquí el ejericio 5.2

# 1. Asegurar la limpieza del DataFrame (eliminar las filas con NaN)
# Es necesario para que train_test_split funcione sin errores.
df_limpio = df.dropna().copy()

# 2. Definir las Variables X (Características) e y (Objetivo)
# X: Todas las columnas excepto la columna objetivo 'species'.
X = df_limpio.drop('species', axis=1)
# y: La columna objetivo 'species'.
y = df_limpio['species']

# 3. Codificar las Variables Categóricas en X (One-Hot Encoding)
# Los modelos de ML (como la Regresión Logística) necesitan números, no strings
# para las columnas 'island' y 'sex'.
X_encoded = pd.get_dummies(X, drop_first=True)

# ----------------------------------------------------------------------------------
# DIVISIÓN DEL DATASET
# ----------------------------------------------------------------------------------

print("--- Realizando la División Estratificada (60% Train, 40% Test) ---")

# 4. Realizar la división estratificada
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y,
    test_size=0.40,
    random_state=42,
    stratify=y  # Mantiene la misma proporción de las 3 especies en ambos conjuntos
)

# --- Verificación ---
print("¡División completada con éxito!")
print(f"\nCaracterísticas (X) tras One-Hot Encoding: {X_encoded.shape[1]}")
print(f"X_train (Entrenamiento): {X_train.shape}")
print(f"X_test (Test/Prueba): {X_test.shape}")

print("\nVerificación de la Proporción de Especies en la División:")
print("Original:\n", y.value_counts(normalize=True).to_markdown(floatfmt=".3f"))
print("\nEntrenamiento:\n", y_train.value_counts(normalize=True).to_markdown(floatfmt=".3f"))
print("\nPrueba:\n", y_test.value_counts(normalize=True).to_markdown(floatfmt=".3f"))
```

#### Ejercicio 5.3
Elige un modelo de clasificación entre los siguientes:
 - Decision tree (Árbol de decisión)
 - Random forest
 - SVM (Support Vector Machine)

Entrenalo con los conjuntos de datos separados en el ejercicio anterior.

```python
# Resuelve aquí el ejericio 5.3


# ----------------------------------------------------------------------------------
# PASO 1: Estandarización (CRUCIAL para SVM y buena práctica)
# ----------------------------------------------------------------------------------
print("--- 1. Estandarización de las Características ---")
scaler = StandardScaler()

# Ajustar y transformar X_train
X_train_scaled = scaler.fit_transform(X_train)

# Transformar X_test (usando los parámetros aprendidos en X_train)
X_test_scaled = scaler.transform(X_test)
print("Características estandarizadas para el entrenamiento y la prueba.")

# ----------------------------------------------------------------------------------
# PASO 2: Entrenamiento y Evaluación de los 3 Modelos
# ----------------------------------------------------------------------------------

# Diccionario para almacenar modelos y resultados
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42, n_estimators=100),
    "Support Vector Machine (SVM)": SVC(random_state=42)
}

results = {}

print("\n--- 2. Entrenamiento y Evaluación en el Conjunto de Prueba ---")
for name, model in models.items():
    # Entrenar el modelo
    model.fit(X_train_scaled, y_train)

    # Generar predicciones en el conjunto de prueba
    y_pred = model.predict(X_test_scaled)

    # Calcular la precisión (Accuracy)
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = accuracy

    print(f"Modelo: {name:<30} | Precisión (Accuracy): {accuracy:.4f}")

# ----------------------------------------------------------------------------------
# PASO 3: Determinar el Mejor Modelo
# ----------------------------------------------------------------------------------

best_model_name = max(results, key=results.get)
best_accuracy = results[best_model_name]

print(f"\n==================================================================")
print(f"🥇 El **MEJOR** modelo de clasificación es: **{best_model_name}**")
print(f"       Con una precisión en el conjunto de prueba de: **{best_accuracy:.4f}**")
print(f"==================================================================")
```
